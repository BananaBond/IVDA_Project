```python
import pandas as pd
import json
import matplotlib.pyplot as plt
from datetime import datetime
import numpy as np
import os
from scipy.signal import savgol_filter
import warnings
warnings.filterwarnings('ignore')
```

```python
score_file_path = 'data/scores.csv'

def read_files(path):
    return pd.read_csv(path)

scores = read_files(score_file_path)
```

```python
scores['type'] = scores['number'].apply(lambda i: "depress" if "condition" in i else "nondepress")
```

```python
# Default Graph
def get_count(attribute):
    df = scores.groupby([attribute,'type'], as_index = False)['number'].count().reset_index(drop=True)
    data = {f"{attribute}_group":eval(df.to_json(orient = 'records'))}
    
            
    with open(f"data/detailgraph/all_{attribute}.json", "w") as outfile:
        json.dump(data, outfile)
    return df

get_count('age')
get_count('gender')
```

```python
# Default Graph
attr_list = ["afftype","edu","marriage","work"]
def get_datainfo(attribute):
    df = depress[["number","madrs1","madrs2",attribute]]
    
    data = {attribute:eval(df.to_json(orient = 'records'))}
            
    with open(f"data/detailgraph/after_{attribute}.json", "w") as outfile:
        json.dump(data, outfile)
    return data

for attr in attr_list:
    get_datainfo(attr)
```
